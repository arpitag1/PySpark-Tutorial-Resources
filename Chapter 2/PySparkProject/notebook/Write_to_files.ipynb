```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master("spark://spark-master:7077") \
    .appName("SalesDataAnalysis") \
    .getOrCreate()
```

```python
df_csv = spark.read.csv("/opt/spark-data/Sales/customers.csv", header=True, inferSchema=True)
```

```python
df_csv.show()
```

```python
df_Orders = spark.read.csv("/opt/spark-data/Sales/orders.csv", header=True, inferSchema=True)
```

```python
df_Orders.show()
```

```python
df_Orders.write.csv("/opt/spark-data/output", header=True, mode='overwrite')
```

```python
df_Orders.write.parquet("/opt/spark-data/output", mode='overwrite')
```

```python
df_Orders.write.json("/opt/spark-data/output", mode='overwrite')
```
